```python
# Install the Ultralytics package
!pip install ultralytics -q

# Import necessary libraries
import os
import yaml
from ultralytics import YOLO
from IPython.display import Image

# Define the dataset configuration
# NOTE: Ensure this path matches your Kaggle dataset's location
dataset_path = '/kaggle/input/hackforindia/HackByte_Dataset'

yolo_config = {
    'train': os.path.join(dataset_path, 'data/train/images'),
    'val': os.path.join(dataset_path, 'data/val/images'),
    'test': os.path.join(dataset_path, 'data/test/images'),
    'nc': 3,
    'names': ['FireExtinguisher', 'ToolBox', 'OxygenTank']
}

# Write the configuration to a YAML file
with open('yolo_params.yaml', 'w') as file:
    yaml.dump(yolo_config, file, default_flow_style=False)

print("âœ… yolo_params.yaml created successfully.")
```

```python
from ultralytics import YOLO
import os

# --- 1. Load the Base Model ---
# We use the standard YOLOv8n model pre-trained on the COCO dataset, as specified in the paper[cite: 79, 81].
model = YOLO('yolov8n.pt')

print("Model loaded. Starting fine-tuning with Freeze=10...")

# --- 2. Fine-Tune the Model ---
# The 'freeze=10' parameter freezes the first 10 layers (0-9) of the backbone.
# Layers 10 through 21 and the detection head will be trained on your dataset.
# Other parameters like epochs and optimizer are set to match the paper's methodology[cite: 131].
results = model.train(
    data='yolo_params.yaml',
    epochs=100,
    imgsz=640,
    freeze=10,
    optimizer='AdamW',
    project='detection_runs',
    name='finetune_freeze_10'
)

print("âœ… Fine-tuning with Freeze=10 complete.")
print(f"Results saved to: {results.save_dir}")
```

```python
from ultralytics import YOLO
import os

# --- 1. Load the STRONGER Base Model ---
# We now use 'yolov8s.pt' for its higher capacity, providing a better starting point for fine-tuning.
model = YOLO('yolov8s.pt')

print("Model loaded. Starting fine-tuning with Freeze=10 and Data Augmentation...")

# --- 2. Fine-Tune the Model with Augmentation ---
# The 'freeze=10' parameter preserves the early feature detectors.
# Targeted augmentation parameters are added to improve robustness.
results = model.train(
    data='yolo_params.yaml',
    epochs=100,
    imgsz=640,
    freeze=10,
    optimizer='AdamW',
    project='detection_runs',
    name='finetune_freeze_10_augmented',
    # --- Added augmentation arguments ---
    degrees=20.0,      # Random rotation
    translate=0.1,     # Random translation
    scale=0.5,         # Random scaling (zoom)
    fliplr=0.5,        # Random horizontal flip
    hsv_v=0.4,         # Adjust brightness/value for varied lighting
    copy_paste=0.1     # Create realistic occlusions
)

print("âœ… Fine-tuning with Freeze=10 and Augmentation complete.")
print(f"Results saved to: {results.save_dir}")
```
